# Image resizing and preprocessing

```python
import os
from PIL import Image

# 원본 이미지가 저장된 디렉토리 경로
input_dir = '/content/drive/MyDrive/Yolov7/images/train'
# 변환된 이미지가 저장될 디렉토리 경로
output_dir = '/content/drive/MyDrive/Yolov7/images/train_resize'

# output_dir 디렉토리가 존재하지 않으면 생성합니다.
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# input_dir 내부의 모든 폴더(각 클래스)를 순회합니다.
for class_folder in os.listdir(input_dir):
    class_path = os.path.join(input_dir, class_folder)
    if os.path.isdir(class_path):  # 디렉토리인지 확인
        output_class_path = os.path.join(output_dir, class_folder)
        if not os.path.exists(output_class_path):
            os.makedirs(output_class_path)

        # 각 클래스 폴더 내의 모든 이미지 파일을 순회합니다.
        for image_name in os.listdir(class_path):
            image_path = os.path.join(class_path, image_name)
            if os.path.isfile(image_path):  # 파일인지 확인
                try:
                    with Image.open(image_path) as img:
                        # 이미지를 RGB로 변환합니다.
                        img_rgb = img.convert('RGB')
                        # 이미지를 64x64 크기로 조정합니다.
                        img_resized = img_rgb.resize((64, 64))
                        # 변환된 이미지를 저장합니다.
                        output_image_path = os.path.join(output_class_path, image_name)
                        img_resized.save(output_image_path)
                        print(f"Processed {output_image_path}")
                except Exception as e:
                    print(f"Error processing {image_path}: {e}")

print("이미지 변환 및 크기 조정이 완료되었습니다.")
```

- Images Rescaling

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 이미지 픽셀을 0 ~ 1 사이의 값으로 스케일 조정
train_dategen = ImageDataGenerator(rescale = 1./255)
test_datagen = ImageDataGenerator(rescale = 1./255)
validation_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
train_generator = train_dategen.flow_from_directory(train_dir,
                                                    batch_size = 32,
                                                    target_size = (128, 128),
                                                    class_mode = 'sparse')

test_generator = test_datagen.flow_from_directory(test_dir,
                                                    batch_size = 32,
                                                    target_size = (128, 128),
                                                    class_mode = 'sparse')

validation_generator = validation_datagen.flow_from_directory(val_dir,
                                                    batch_size = 32,
                                                    target_size = (128, 128),
                                                    class_mode = 'sparse')
```

```python
# 클래스 확인
train_generator.class_indices
```

#  Model Configuration

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential()

model.add(Conv2D(filters = 16, kernel_size = 1, padding = 'same',
                 activation = 'relu', input_shape = (128, 128, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))     # 드롭아웃 추가

model.add(Conv2D(filters = 32, kernel_size = 1, padding = 'same',
                 activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))     # 드롭아웃 추가

model.add(Flatten())
model.add(Dense(256, activation = 'relu'))
model.add(Dense(11, activation = 'softmax')) # 다중 분류 -> 출력 11개

# EarlyStopping 콜백 정의
early_stopping = EarlyStopping(monitor='val_loss', patience=45,
                               restore_best_weights=True)

# 모델 컴파일
model.compile(
    optimizer=Adam(1e-4),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# 모델 요약 출력
model.summary()
```

 - 모델 학습

```python
history = model.fit(train_generator,
                    epochs = 200,
                    batch_size = 16,
                    validation_data = validation_generator,
                    callbacks= [early_stopping])
```

 - 학습과정 그려보기

```python
import matplotlib.pyplot as plt

his_dict = history.history
loss = his_dict['loss']
val_loss = his_dict['val_loss']

epochs = range(1, len(loss) + 1)
fig = plt.figure(figsize = (10, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, color = 'blue', label = 'train_loss')
ax1.plot(epochs, val_loss, color = 'orange', label = 'val_loss')
ax1.set_title('train and val loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('loss')
ax1.legend()

acc = his_dict['accuracy']
val_acc = his_dict['val_accuracy']

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, acc, color = 'blue', label = 'train_acc')
ax2.plot(epochs, val_acc, color = 'orange', label = 'val_acc')
ax2.set_title('train and val acc')
ax2.set_xlabel('epochs')
ax2.set_ylabel('acc')
ax2.legend()

plt.show()
```

# CNN Feature Extractor

```python
import numpy as np
import os
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.models import Model

# VGG16 모델 로드 및 FC 레이어 제거
base_model = VGG16(weights='imagenet', include_top=False)

# 특성 추출을 위한 새로운 모델 정의 (block5_conv3 레이어의 출력을 특성으로 사용)
model = Model(inputs=base_model.input, outputs=base_model.get_layer('block5_conv3').output)
```

```python
def extract_features(img_path):
    # 이미지 로드 및 전처리
    img = image.load_img(img_path, target_size=(64, 64))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    # 특성 추출
    features = model.predict(x)
    return features
```

```python
def process_and_save_features(data_dir, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 데이터셋 종류별 폴더 순회 (train, test, val)
    for dataset_type in ['train', 'test', 'val']:
        dataset_path = os.path.join(data_dir, dataset_type)
        output_dataset_path = os.path.join(output_dir, dataset_type)

        if not os.path.exists(output_dataset_path):
            os.makedirs(output_dataset_path)

        # 질환 폴더 순회
        for disease_folder in os.listdir(dataset_path):
            disease_path = os.path.join(dataset_path, disease_folder)
            if os.path.isdir(disease_path):
                features_list = []
                image_files = [f for f in os.listdir(disease_path) if f.endswith(('.jpg', '.jpeg', '.png'))]

                # 각 이미지 파일에 대해 특성 추출
                for img_file in image_files:
                    img_path = os.path.join(disease_path, img_file)
                    features = extract_features(img_path)
                    features_list.append(features)

                # 특성을 numpy 배열로 변환 및 저장
                features_array = np.vstack(features_list)
                output_file = os.path.join(output_dataset_path, f"{disease_folder}_features.npy")
                np.save(output_file, features_array)
                print(f"Features saved to {output_file}")
```

```python
# 데이터 디렉토리 경로와 출력 디렉토리 경로 설정
data_dir = '/content/drive/MyDrive/xgboost/images'  # 이미지가 저장된 폴더 경로 (train, test, val 폴더가 포함된 상위 경로)
output_dir = '/content/drive/MyDrive/예비프로젝트_천대원/CNN/feature_extract'  # 특성 파일을 저장할 경로

# 특성 추출 및 저장 실행
process_and_save_features(data_dir, output_dir)
```
